```python
import pandas as pd
table = pd.DataFrame(results["search-results"]["entry"])
```

```python
def find_source_link(list_of_links):
    for link in list_of_links:
        if link["@ref"] == "scopus":
            return link["@href"]
```

```python
table["source_link"] = table["link"].apply(lambda x: find_source_link(x))
```

```python
abstract_retieval_api = "https://api.elsevier.com/content/abstract/scopus_id/{scopus_id}"
```

```python
table["scopus_id"] = table["dc:identifier"].apply(lambda x: x.split("SCOPUS_ID:")[1])
```
