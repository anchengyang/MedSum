```python
# gettng the text from a page
response = requests.get("https://medlineplus.gov/genetics/condition/dyserythropoietic-anemia-and-thrombocytopenia/")
response.text
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
content = soup.find_all("div", class_="main")

for c in content:
    list_of_bookmarks = c.find_all("div", class_="mp-exp exp-full")
```

```python
details = {}
for bookmark in list_of_bookmarks:
    header = bookmark.find("h2").text
    text = bookmark.find_all("section")
    res = []
    print(header)
    print(text)
    for t in text:
        if t.find("ul", class_="bulletlist"):
            text = t.find("ul", class_="bulletlist")
            res = [li.text for li in t.find_all("li")]    
        elif t.find("p"):
            res.append(t.find("p").text)
        elif t.find("h2") and t.find("ul"):
            subheader = t.find("h2").text.strip()
    
            # Extract resources
            resources = []
            for link in t.find_all("a"):
                resources.append({"title": link.text.strip(), "url": link.get("href")})
            
            # Append dictionary to result list
            res.append({"subHeader": subheader, "resources": resources})
        elif len(t.find_all("li")) > 1:
            res = [li.text for li in t.find_all("li")]


            
    print(res)
    details[header] = res
```
